```python
from sklearn.model_selection import train_test_split

train_2= train.drop(['id'],axis=1)
train_fit=train_2.drop(['count'],axis=1)
train_trg=train_2['count']
tr_x,vl_x,tr_y,vl_y=train_test_split(train_fit,train_trg,test_size=0.25,shuffle=True,random_state=613)
```

```python
vl_x_scl=pd.DataFrame(vl_x_scl,columns=tr_x.columns)
vl_x_scl.head()
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import KFold,GridSearchCV
from sklearn.metrics import mean_squared_error 
```

```python
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=150,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =7, nthread = -1)
```

```python
# A parameter grid for XGBoost
params = {'min_child_weight':[i/10.0 for i in range(5,18)], 'gamma':[i/100.0 for i in range(3,6)],  
'subsample':[i/10.0 for i in range(4,9)], 'colsample_bytree':[i/10.0 for i in range(4,8)], 'max_depth': [2,3,4]}

# Initialize XGB and GridSearch

grid = GridSearchCV(model_xgb, params)
```

```python
grid.fit(tr_x_scl,tr_y)
```

```python
grid.predict(test_2)
```

```python
np.sqrt(mean_squared_error(vl_y,grid.predict(vl_x_scl)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True, random_state=42)

min_val_error = float("inf")
error_going_up = 0
for n_estimators in range(1,200):
    gbrt.n_estimators = n_estimators
    gbrt.fit(tr_x_scl, tr_y)
    y_pred = gbrt.predict(vl_x_scl)
    val_error = np.sqrt(mean_squared_error(vl_y, y_pred))
    if val_error < min_val_error:
        min_val_error = val_error
        error_going_up = 0
    else:
        error_going_up += 1
        if error_going_up == 5:
            break  # 조기 종료
```

```python
print('최적의 트리 개수 :', gbrt.n_estimators)
print("최소 검증 MSE:", min_val_error)
```

```python
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=85, random_state=613)
gbrt.fit(tr_x_scl, tr_y)
```

```python
# 최적의 트리 개수 찾기
errors = [mean_squared_error(vl_y, y_pred)
          for y_pred in gbrt.staged_predict(vl_x_scl)]
bst_n_estimators = np.argmin(errors)

bst_n_estimators
```

```python
# 최적의 트리개수로 그래디언트 부스팅 학습
gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators=bst_n_estimators, random_state=613)
gbrt_best.fit(tr_x_scl, tr_y)
```

```python
np.sqrt(mean_squared_error(vl_y,gbrt_best.predict(vl_x_scl)))
```

```python
# 1번 2번  XGBRegressor() 생성
model=xgb.XGBRegressor()
param_grid={'booster' :['gbtree'],
                 'silent':[True],
                 'max_depth':[5,6,8],
                 'min_child_weight':[1,3,5,7,9],
                 'gamma':[i/100.0 for i in range(3,6)],
                 'nthread':[-1],
                 'colsample_bytree':[0.5,0.8,0.9],
                 'colsample_bylevel':[0.9],
                 'learning_rate':[0.05,0.07,0.1],
                 'n_estimators':[65,85,119,115,120],
                 'objective':['reg:linear'],
                 'random_state':[12]}
# 3번  KFold 지정
cv=KFold(n_splits=10, random_state=11)

# 4번  GridSearchCV로 모델 생성
gcv=GridSearchCV(model, param_grid=param_grid, cv=cv, scoring='neg_mean_squared_error', n_jobs=-1)
```

```python
# 6번  최적의 파라미터로 xgboost 모델 생성
model= gcv.best_estimator_
model
```
